Phase 4: Data Manipulation & Cleaning
• Task 9: Fill missing Age values with the median.

Phase 5: Data Visualization
• Task 13: Plot histogram of Age.
• Task 14: Plot bar chart of Survival Count.

```python
print("Maximum Fare:", np.max(fare_array))
print("Average Fare:", np.mean(fare_array))
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Fare"], color="red")
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
